```python
# Merging train and test data for feature engineering
ntrain = data.shape[0]
ntest = test.shape[0]

all_data = pd.concat((data,test)).reset_index(drop = True)
```

```python
# To fit model using xgboost
from xgboost import XGBClassifier 
```

```python
model = XGBClassifier()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop('mobile_money_classification', axis = 1)
y = data['mobile_money_classification']
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
pre = model.predict_proba(X_test)
```

```python
from sklearn.metrics import log_loss
```

```python
log_loss(y_test, pre)
```
